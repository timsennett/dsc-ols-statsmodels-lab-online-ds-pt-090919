## Step 1: Read the dataset and inspect its columns and 5-point statistics

```python
df = pd.read_csv('advertising.csv')
```

```python
# Check the columns and first few rows
df.head()
```

```python
# Get the 5-point statistics for data 
df.describe()
```

```python
# Describe the contents of this dataset

# Looking at the mean and median of each feature, it appears there
# is probably an outlier skewing the newspaper data, and possibly, to a
# lesser extent, the sales data as well.
```


## Step 3: Test for the linearity assumption 

Use scatterplots to plot each predictor against the target variable

```python
# visualize the relationship between the predictors and the target using scatterplots

# I'm going to make a subplot with four different plots in it.
# That means I need 1 figure and 4 axes. (1, 4, 1) means 1 row, 4 columns, subplot 1

fig = plt.figure(figsize = (18,6))
ax1 = plt.subplot(1, 4, 1)
ax2 = plt.subplot(1, 4, 2)
ax3 = plt.subplot(1, 4, 3)
ax4 = plt.subplot(1, 4, 4)

ax1.scatter(df['Unnamed: 0'], df['sales'])
ax1.set_title('Index vs Sales')
ax2.scatter(df['TV'], df['sales'])
ax2.set_title('TV vs Sales')
ax3.scatter(df['radio'], df['sales'])
ax3.set_title('Radio vs Sales')
ax4.scatter(df['newspaper'], df['sales'])
ax4.set_title('Newspaper vs Sales')
plt.show()
```

```python
# Record yor observations on linearity here 

# The strongest relationship belongs to TV and Sales.
# Clearly TV is going to be our strongest predictor of Sales.
# Radio also appears to have a positive relationship to Sales, albiet a weaker one.
# Newspaper and Sales do not have a linear relationship.
```

### Conclusion so far

Based on above initial checks, we can confidently say that TV and radio appear to be good predictors for our regression analysis. Newspaper is very heavily skewed and also doesnt show any clear linear relationship with the target. 
> We'll move ahead with our analysis using TV and radio, and rule out newspaper because we believe it violates OLS assumptions

Note: Kurtosis can be dealt with using techniques like log normalization to "push" the peak towards the center of distribution. You'll learn about this later on.

## Step 4: Run a simple regression in Statsmodels with TV as a predictor

```python
# import libraries
# I already imported it at the top of the lab

# build the formula 
f = 'sales~TV'

# create a fitted model in one line
model = ols(formula = f, data = df).fit()
```

## Step 5: Get Regression Diagnostics Summary

```python
model.summary()
```

Note here that the coefficients represent associations, not causations

## Step 6:  Draw a prediction line with data points on a scatter plot for X (TV) and Y (Sales)

Hint: You can use the `model.predict()` function to predict the start and end point of of regression line for the minimum and maximum values in the 'TV' variable. 

```python
# create a DataFrame with the minimum and maximum values of TV
TVminmax = pd.DataFrame({'TV': [df['TV'].min(), df['TV'].max()]})

# make predictions for those x values and store them
preds = model.predict(TVminmax)
preds

# first, plot the observed data and the least squares line

df.plot(x = 'TV', y = 'sales', kind = 'scatter')
plt.plot(TVminmax, preds, c = 'gray')
plt.show()
```

## Step 7: Visualize the error term for variance and heteroscedasticity

```python
fig = plt.figure(figsize=(15,8))
fig = sm.graphics.plot_regress_exog(model, "TV", fig=fig)
plt.show()
```

```python
# Record Your observations on heteroscedasticity

# The data points are heteroscedastic, as the y variable (sales)
# widens as the x-variable increases, resulting in a cone-like shape.
```

## Step 8: Check the normality assumptions by creating a QQ-plot

```python
# Code for QQ-plot here

import scipy.stats as stats
residuals = model.resid
fig = sm.graphics.qqplot(residuals, dist=stats.norm, line='45', fit=True)
fig.show()
```

```python
# Record Your observations on the normality assumption


# While the residuals do not seem to match up perfectly with the red line, there seem to be no dramatic
# deviations from the red line. It can be assumed that we're OK for the normality assumption.
```

## Step 9: Repeat the above for radio and record your observations

```python
# code for model, prediction line plot, heteroscedasticity check and QQ normality check here
```

```python
f2 = 'sales~radio'
```

```python
model2 = ols(formula = f2, data = df).fit()
```

```python
model2.summary()
```

```python
radio_minmax = pd.DataFrame({'radio': [df['radio'].min(), df['radio'].max()]})

preds2 = model2.predict(radio_minmax)
preds2

df.plot(x = 'radio', y = 'sales', kind = 'scatter')
plt.plot(radio_minmax, preds2, c = 'gray')
plt.show()
```

```python
fig = plt.figure(figsize=(15,8))
fig = sm.graphics.plot_regress_exog(model2, "radio", fig=fig)
plt.show()
```

```python
residuals = model2.resid
fig = sm.graphics.qqplot(residuals, dist=stats.norm, line='45', fit=True)
fig.show()
```

```python
# Record your observations here for goodnes of fit

# Right away, it looked like radio wasn't going to be as reliable a predictor of sales as TV.
# This was seen by the lower R-squared value.

# In the Q-Q Plot, there are significant deviations on both ends of the spectrum.

# The data does not pass the homoscadascticity assumption, judged from the cone-like shape
# resulting from the y-value of sales spreading high and low as the x value  of radio increases.
```

## The Answer

There is obvious heteroscdasticity in the residuals for both variables.
